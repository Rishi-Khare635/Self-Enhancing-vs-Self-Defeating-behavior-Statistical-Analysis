```python
import pandas as pd

df = pd.read_csv("../data/data.csv")

print(df.shape)
print(df.head())

# 32 questions, 7 different categories tested/calculated
```

```python
# We have found an error, there are a lot of people with Q scores of -1. when computing the specific scores for 
# type of humor with this "-1", we get a completely different result from the scores noted in the original CSV file.
# We will determine how big this error is with this cell by seeing how many -1s are there in each row from Q1-Q32
q_columns = [f'Q{i}' for i in range(1, 33)]
rows_with_missing = (df[q_columns] == -1).any(axis=1).sum()
print(f"Rows with at least one missing answer: {rows_with_missing}")
print(f"Total rows: {len(df)}")


# we find out that out of 1071 rows, there are 78 rows with at least one "-1" score. we decide to remove those rows as 993
# proper rows is still good enough for us
```

```python
# this is how you remove 

has_missing = (df[q_columns] == -1).any(axis =1) # gives the lists of all 1071 trues and false if -1 exists. true means -1 exists
df_clean = df[~has_missing] # this puts all rows that do NOT have a -1, filtered rows
print(df_clean.shape)
```

```python
# Now that we removed all the confusion, we can use the formula mentioned in the codebook.

#affiliative. round(((6-$_POST['Q1']) + $_POST['Q5'] + (6-$_POST['Q9']) + $_POST['Q13'] + (6-$_POST['Q17']) + $_POST['Q21'] + (6-$_POST['Q25']) + (6-$_POST['Q29']))/8, 1);
#selfenhancing. round(($_POST['Q2'] + $_POST['Q6'] + $_POST['Q10'] + $_POST['Q14'] + $_POST['Q18'] + $_POST['Q22'] + $_POST['Q26'] + $_POST['Q30'])/8,1);
#aggressive. round(($_POST['Q3']+ $_POST['Q7'] + $_POST['Q11'] + $_POST['Q15'] + $_POST['Q19'] + $_POST['Q23'] + $_POST['Q27'] + $_POST['Q31'])/8,1);
#selfdefeating. round(($_POST['Q4'] + $_POST['Q8'] + $_POST['Q12'] + $_POST['Q16'] + $_POST['Q20'] + $_POST['Q24'] + $_POST['Q28'] + $_POST['Q32'])/8,1);

df_clean['affiliative'] = ((6-df_clean['Q1']) + df_clean['Q5'] + (6-df_clean['Q9']) + df_clean['Q13'] + (6-df_clean['Q17']) + df_clean['Q21'] + (6-df_clean['Q25']) + (6-df_clean['Q29']))/8
df_clean['selfenhancing'] = (df_clean['Q2'] + df_clean['Q6'] + df_clean['Q10'] + df_clean['Q14'] + df_clean['Q18'] + df_clean['Q22'] + df_clean['Q26'] + df_clean['Q30'])/8
df_clean['aggressive'] = (df_clean['Q3'] + df_clean['Q7'] + df_clean['Q11'] + df_clean['Q15'] + df_clean['Q19'] + df_clean['Q23'] + df_clean['Q27'] + df_clean['Q31'])/8
df_clean['selfdefeating'] = (df_clean['Q4'] + df_clean['Q8'] + df_clean['Q12'] + df_clean['Q16'] + df_clean['Q20'] + df_clean['Q24'] + df_clean['Q28'] + df_clean['Q32'])/8

print(df_clean[['affiliative', 'selfenhancing', 'aggressive', 'selfdefeating']].head())

# we still have some of the original data, to not confuse ourselves, we will drop the columns that are not needed

print(df_clean.columns.tolist()) 
#df_clean.columns.tolist()
#df_clean.drop(columns= ['agressive'])
#df_clean = df_clean.drop(columns=['agressive'])

print(df_clean.columns.tolist())  # removes the pre-made column called 'agressive' from the original CSV file
```

```python
df_clean.to_csv('../data/clear_humor_data.csv', index = False)
print("Clean data saved successfully")
```
